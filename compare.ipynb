```python
import numpy as np
import matplotlib.pyplot as plt

# 读两个文件
data1 = np.loadtxt("/home/junrui/Ptest/Spyral-1.0/spyral_notebooks/spectrum_0.18.txt")
data2 = np.loadtxt("/home/junrui/Ptest/Spyral-1.0/spyral_notebooks/spectrum_1.0.txt")
%matplotlib widget
# 第一个
c1 = 0.5 * (data1[:,0] + data1[:,1])
y1 = data1[:,2]

# 第二个
c2 = 0.5 * (data2[:,0] + data2[:,1])
y2 = data2[:,2]

# 画图
fig, ax = plt.subplots()

ax.step(c1, y1, where="mid", color="black", label="0.18")
ax.step(c2, y2, where="mid", color="red", label="1.0")

ax.set_xlabel("Excitation Energy (MeV)")
ax.set_ylabel("Counts")
ax.legend()

plt.show()
```
